1) Find Maximum Using Loop and Condition Only

```r
x <- c(4, 7, 2, 9, 5)

max_val <- x[1]

for (i in x) {
  if (i > max_val) {
    max_val <- i
  }
}

max_val
```

2) Replace NA with Mean (Without Using mean())

```r

x <- c(10, NA, 20, NA, 30)

sum <- 0
count <- 0

for (i in x) {
  if (!is.na(i)) {
    sum <- sum + i
    count <- count + 1
  }
}

mean_value <- sum / count

for (i in 1:length(x)) {
  if (is.na(x[i])) {
    x[i] <- mean_value
  }
}

x
```

3) Frequency of Each Unique Element Using Loops

```r

x <- c(2, 3, 2, 5, 3, 2)

unique_vals <- c()

for (i in x) {
  if (!(i %in% unique_vals)) {
    unique_vals <- c(unique_vals, i)
  }
}

for (u in unique_vals) {
  count <- 0
  for (i in x) {
    if (i == u) {
      count <- count + 1
    }
  }
  cat("Element:", u, "Frequency:", count, "\n")
}
```

4) Find the second largest element in a vector without sorting and without using
built-in functions.

```r
x <- c(10, 25, 5, 30, 20)

largest <- -Inf
second_largest <- -Inf

for (i in x) {
  if (i > largest) {
    second_largest <- largest
    largest <- i
  } else if (i > second_largest && i != largest) {
    second_largest <- i
  }
}

second_largest
```

5) Check whether a vector is a palindrome (without using rev() or built-in
functions)

```r
x <- c(1, 2, 3, 2, 1)

is_palindrome <- TRUE
n <- length(x)

for (i in 1:(n %/% 2)) {
  if (x[i] != x[n - i + 1]) {
    is_palindrome <- FALSE
    break
  }
}

if (is_palindrome) {
  cat("The vector is a palindrome")
} else {
  cat("The vector is NOT a palindrome")
}
```

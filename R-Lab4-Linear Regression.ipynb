```r
# Install package (only once)
install.packages("MASS")

# Load library
library(MASS)

# Load dataset
data(Boston)

# View dataset
head(Boston)

# Structure
str(Boston)


# Summary
summary(Boston)
```

 Dependent Variable (Y):
 medv -> Median value of owner-occupied homes (House Price)

 Independent Variables (X):
 crim, zn, indus, chas, nox, rm, age,
 dis, rad, tax, ptratio, black, lstat

**STEP 2: Check Missing Values**

```r
colSums(is.na(Boston))
```

**STEP 3: Exploratory Data Analysis (EDA)**

Correlation Matrix

```r
cor_matrix <- cor(Boston)
round(cor_matrix, 2)
```

Scatter Plot (medv vs lstat)

```r
plot(Boston$lstat, Boston$medv,
     main="House Price vs Lower Status Population",
     xlab="lstat",
     ylab="medv",
     col="blue",
     pch=19)
```

**STEP 4: Train-Test Split (80-20)**

```r
set.seed(123)

sample_index <- sample(1:nrow(Boston), 0.8 * nrow(Boston))

train <- Boston[sample_index, ]
test  <- Boston[-sample_index, ]
```

**STEP 5: Build Multiple Linear Regression Model**

```r
model <- lm(medv ~ ., data = train)

summary(model)
```

**STEP 6: Prediction on Test Data**

```r
predictions <- predict(model, test)

head(predictions)
```

**STEP 7: Model Evaluation**

```r
rmse <- sqrt(mean((test$medv - predictions)^2))
rmse
```

```r
sst <- sum((test$medv - mean(test$medv))^2)
sse <- sum((test$medv - predictions)^2)

r2_test <- 1 - (sse/sst)
r2_test
```

**STEP 9: Simple Linear Regression**

```r
simple_model <- lm(medv ~ lstat, data=train)

summary(simple_model)

plot(train$lstat, train$medv,
     col="blue", pch=19,
     main="Simple Linear Regression",
     xlab="lstat",
     ylab="medv")

abline(simple_model, col="red", lwd=2)
```
